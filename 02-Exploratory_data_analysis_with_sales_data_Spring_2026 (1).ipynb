### Question 1: Import Libraries and Load Data (5 Marks)

Import the `pandas` library as `pd`. Then, download the data `Coffe_sales.csv` from https://drive.google.com/drive/folders/1TeLp4U4NsXCSgClbF7ODBsaLKpHSWeQr?usp=sharing and load into a DataFrame named `coffee_data` and display its first 5 rows.

**Hint:** Use `pd.read_csv()` to load the data. To display the first few rows, use the `.head()` method.

**Expected Output:** A table showing the first 5 rows of the coffee sales data.

```python
import pandas as pd
from google.colab import drive


coffee_data = pd.read_csv('/content/drive/MyDrive/Coffe_sales.csv')

coffee_data.head()


```

### Question 2: Initial Data Inspection (5 Marks)

Perform a basic inspection of the `coffee_data` DataFrame. Print the number of rows and columns, check for any duplicate rows, and find the total number of missing values in the entire dataset.

**Hint:** Use `.shape` for dimensions, `.duplicated().sum()` for duplicate count, and `.isnull().sum().sum()` for total missing values.

**Expected Output:** A printout of the shape, the number of duplicate rows, and the total number of missing values.

```python

print("Shape of the dataset (rows, columns):", coffee_data.shape)

duplicate_rows = coffee_data.duplicated().sum()
print("Number of duplicate rows:", duplicate_rows)


missing_values = coffee_data.isnull().sum().sum()
print("Total number of missing values:", missing_values)
```

### Question 4: Data Type Conversion (10 Marks)

The 'Date' column is currently an object. Convert it to a datetime object. Then, create two new columns: 'Month' (containing the month number) and 'Year' (containing the year as an integer). Finally, display the data types of all columns.

**Hint:** Use `pd.to_datetime()` to convert the 'Date' column. Access month and year using `.dt.month` and `.dt.year` on the datetime column. Use `.dtypes` to check the final data types.

**Expected Output:** A series listing each column and its updated data type.

```python
coffee_data['Date'] = pd.to_datetime(coffee_data['Date'])

coffee_data['Month'] = coffee_data['Date'].dt.month
coffee_data['Year'] = coffee_data['Date'].dt.year

coffee_data.dtypes
```

### Question 5: Average Sales per Year (10 Marks)

Calculate the average 'money' (sales) for each 'Year'. Group the data by the 'Year' column and then compute the mean of the 'money' column for each group.

**Hint:** Use the `.groupby('Year')['money'].mean()` method.

**Expected Output:** A Pandas Series showing the years as the index and the average sales as the values.

```python

average_sales_per_year = coffee_data.groupby('Year')['money'].mean()

average_sales_per_year
```

### Question 6: Maximum Sales by Month Name (10 Marks)

Find the maximum 'money' earned for each 'Month_name'. Display the result as a Pandas Series.

**Hint:** Use `.groupby('Month_name')['money'].max()`.

**Expected Output:** A Series with month names as the index and their corresponding maximum sales value.

```python

max_sales_by_month = coffee_data.groupby('Month_name')['money'].max()

max_sales_by_month
```

### Question 7: Total Sales by Coffee Name (10 Marks)

Calculate the total 'money' (sum of sales) for each 'coffee_name'.

**Hint:** Use `.groupby('coffee_name')['money'].sum()`.

**Expected Output:** A Series with coffee names as the index and the total sales for each.

```python


total_sales_by_coffee = coffee_data.groupby('coffee_name')['money'].sum()

total_sales_by_coffee
```

### Question 8: Unique Coffee Types (15 Marks)

Determine how many unique types of coffee are present in the dataset. Print the unique coffee names as a list.

**Hint:** Use the `.unique()` method on the 'coffee_name' column and convert the result to a list using `list()`.

**Expected Output:** A Python list containing the names of all unique coffee types.

```python
unique_coffee_types = list(coffee_data['coffee_name'].unique())

print("Number of unique coffee types:", len(unique_coffee_types))

unique_coffee_types
```

### Question 9: Average Sales by Time of Day (15 Marks)

Find the average 'money' made during each 'Time_of_Day' (Morning, Afternoon, Night).

**Hint:** Use `.groupby('Time_of_Day')['money'].mean()`.

**Expected Output:** A Series with 'Time_of_Day' as the index and the average sales for each period.

```python


average_sales_by_time = coffee_data.groupby('Time_of_Day')['money'].mean()

average_sales_by_time
```

### Question 10: Generate and Display Synthetic Data (15 Marks)

Create a new, synthetic DataFrame named `synthetic_df`. It should have two columns: 'date' and 'value'. The 'date' column should contain 100 dates starting from '2023-01-01'. The 'value' column should contain 100 random numbers. Display the entire synthetic DataFrame.

**Hint:** Use `pd.date_range(start="2023-01-01", periods=100)` for the dates and `np.random.randn(100)` for the random values. Combine them into a DataFrame using a dictionary.

**Expected Output:** A 100x2 DataFrame with 'date' and 'value' columns.

```python
import numpy as np
import pandas as pd

dates = pd.date_range(start="2023-01-01", periods=100)

values = np.random.randn(100)

synthetic_df = pd.DataFrame({
    'date': dates,
    'value': values
})

synthetic_df
```
